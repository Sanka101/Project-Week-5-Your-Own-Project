```python
# 368653 Coliseum Alfonso Perez 2-0 , 368654  Arena 2-1
fixtures_uefa=[368653,368654]
```

```python
full_response= []


for fixture in range(0,2):
    url = "https://v2.api-football.com/statistics/fixture/" +str(fixtures_uefa[fixture])

    payload = {}
    headers = {
      'x-rapidapi-key': '85fcf85753adf7519cc4f7b7535edf10'
    }

    response = requests.request("GET", url, headers=headers, data = payload).json()
    
    # Add dict of the response in our full_response list
    full_response.append(response)
```

```python
full_response
```

```python
stats_list =[]
for i in range(0,2):
    stats_list.append(full_response[i]['api']['statistics'])
stats_list 
```

```python
super_list= []
for i in range(0,2):
    a,b = zip(*stats_list[i].items())
    a = list(a)
    b = list(b)
    super_list.append(b)

super_list
```

```python
home_getafe= ['Home','Away']
```

```python
clean_stats_uefa= []

for i in range(0,2):
    if home_getafe[i] == 'Home':
        for j in range(0,16):
            clean_stats_uefa.append(super_list[i][j]['home'])
    elif home_getafe[i] == 'Away':
        for k in range(0,16):
            clean_stats_uefa.append(super_list[i][k]['away'])
        
clean_stats_uefa
```

```python
for i in range(0, len(clean_stats_uefa), 16):
    chunk = clean_stats_uefa[i:i + 16]
    print(chunk)
```
