```python
# From result_get, we want to extract all ficxture_id 's/matches in La Liga
# List of  dict, all 38 fixtures

fixtures_ajax= []
fixtures_aja_home = []
for game in range(0,34):
    fixtures_ajax.append(fixtures_ajax_all['api']['fixtures'][game]['fixture_id'])
    fixtures_aja_home.append(fixtures_ajax_all['api']['fixtures'][game]['homeTeam'])

print(len(fixtures_ajax))
fixtures_aja_home
```

```python
full_response= []


for fixture in range(0,34):
    url = "https://v2.api-football.com/statistics/fixture/" +str(fixtures_ajax[fixture])

    payload = {}
    headers = {
      'x-rapidapi-key': '85fcf85753adf7519cc4f7b7535edf10'
    }

    response = requests.request("GET", url, headers=headers, data = payload).json()
    
    # Add dict of the response in our full_response list
    full_response.append(response)
```
