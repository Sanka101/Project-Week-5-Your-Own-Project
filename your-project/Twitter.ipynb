```python
import twitterscraper
import pandas as pd 
import numpy as np
```

```python
from twitterscraper import query_tweets
import datetime as dt
```

```python
begin_date = dt.date(2019, 9,10)
end_date = dt.date(2020,7, 28)
limit= 10000
# language
#user = AFCAjax
```

```python
# you can add dates
ajax_tweets = query_tweets("from:AFCAjax", begindate= begin_date, enddate = end_date, 
                           limit =limit,
                          )
```

```python
df = pd.DataFrame(t.__dict__ for t in ajax_tweets)
```

```python
df
```
