```python
# From result_get, we want to extract all ficxture_id 's/matches in La Liga
# List of  dict, all 38 fixtures

fixtures_geta= []
fixters_home = []
for game in range(0,38):
    fixtures_geta.append(result_get['api']['fixtures'][game]['fixture_id'])
    fixters_home.append(result_get['api']['fixtures'][game]['homeTeam'])

len(fixtures_geta)
fixters_home
```

```python
full_response= []


for fixture in range(0,38):
    url = "https://v2.api-football.com/statistics/fixture/" +str(fixtures_geta[fixture])

    payload = {}
    headers = {
      'x-rapidapi-key': '85fcf85753adf7519cc4f7b7535edf10'
    }

    response = requests.request("GET", url, headers=headers, data = payload).json()
    
    # Add dict of the response in our full_response list
    full_response.append(response)
```
